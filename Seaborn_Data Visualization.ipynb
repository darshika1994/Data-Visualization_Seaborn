```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('white')
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

## 1. scatter plot - both continuous variables

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df)
plt.show()
```

## 2. scatterplot - two continuous variables + a categorical variable (hue) - SHOWING ERROR

#### adding hue

```python
sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df, hue = df['species'].tolist())
plt.show()
```

#### adding hue order and hue color

```python
hue_colors = {'Gentoo':'black','Chinstrap':'red','Adelie':'blue'}

sns.scatterplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = df, hue = df['species'].tolist(),
hue_order = ['Gentoo','Chinstrap','Adelie'], palette = hue_colors)

plt.show()
```

## 3. count plot - one categorical variable

```python
sns.countplot(x = 'species', data=df)
```

## 4. count plot - one categorical variable + a categorical variable (hue)

```python
sns.countplot(x = 'species', data=df, hue = 'island')
```

## 5. relplot() - 
### relational plots helps to identify relationship between two quantitaive variable either using a scatter plot or a line plot. It is similar to scatter plot but it lets us create subplots in a single figure. Specify the kind (scatter or line)

#### adding a col (which is just like hue) but in relational plot

```python
sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', col = 'species')
```

#### adding a col_wrap

```python
sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', col = 'species', col_wrap = 2)
```

#### adding a col_wrap + col_order

```python
sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', col = 'species', col_wrap = 2,
           col_order = ['Gentoo','Chinstrap','Adelie'])
```

#### adding a row (which is just like hue) but in relational plot (will display result in separate row)

```python
sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', row = 'species')
```

#### adding both col and row (which are just like hue) but in relational plot

```python
sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', col = 'species', row = 'island')
```

## 8. Box plot

#### shows districution of quantitaive data. whiskers gives a sense of distribution, and floating points represent outliers. They are used to compare the distribution of a quantitative variable across different groups of a categorical variable. Put categorical variable on x axis and quantitaive variable on y axis.

```python
sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df)
# biscor island has the largest median of body_mass_g as compared to other categories
```

###### changing order 

```python
sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df, order = ['Biscoe','Torgersen','Dream'])
```

##### omitting outlier from box plot - use sym = "" for it

```python
sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df, order = ['Biscoe','Torgersen','Dream'], sym = "")
```

#### whisker (whis) is by default 1.5*IQR, we can change it to 2*IQR by setting whis=2.0. For showing just 5th and 95th percentile, set whis = [5,95] where 5th percentile corresponds to lower whisker and 95th percentile corresponds to upper whisker.Show min and max values whis = [0,100]

```python
sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df, whis = [0,100])
# no outlier could be seen as the whisker cover the entire range of the data
```

## 9. Point Plots - type of categorical plot

#### Point plot show the mean of quantitative variable for the observation in each category, plotted as single plot. x axis will have categorical feature and y axis will have continuous feature

##### Both line plot and point plot shows mean of quantitative variable and 95% confidence interval for the mean. But in line plot both axis have quantitative values whereas in point plot, the x axis is categorical in nature

##### Point plot and bar plot are same, that is it also shows mean of quantitative variable and 95% confidence interval for the mean. But in some cases, point plot shows clear difference as compared to bar plots

```python
sns.catplot(x = 'island', y = 'body_mass_g', data = df, hue = 'sex', kind = 'point')
```

```python
sns.catplot(x = 'island', y = 'body_mass_g', data = df,hue = 'sex', kind = 'bar')
# just comparing point with bar plot and noticing the difference
```

##### setting the join = False

```python
sns.catplot(x = 'island', y = 'body_mass_g', data = df, hue = 'sex', kind = 'point', join = False)
```

##### displaying the median instead of mean(as mean is by default) 

```python
from numpy import median
sns.catplot(x = 'island', y = 'body_mass_g', data = df, hue = 'sex', kind = 'point', estimator = median)
# if our datatset has lots of outlier then median would be better to use instead of mean
```

##### capsize = 0.2

```python
sns.catplot(x = 'island', y = 'body_mass_g', data = df, hue = 'sex', kind = 'point', capsize = 0.2)
```

##### ci = None (ci = confidence interval)

```python
sns.catplot(x = 'island', y = 'body_mass_g', data = df, hue = 'sex', kind = 'point', ci = None)
```

## 11. Adding titles and labels

#### seaborn creates two diff types of objects : FacetGrid (can create subplot) and AxesSubplot (can create single plot)
#### relplot() and catplot() - FacetGrid 
#### scatterplot() and countplot() - AxesSubplot
FacetGrid has one or many axessubplot

### 1. assigning a title to FacetGrid object(relplot() or catplot()) - use g.fig.suptitle

```python
sns.set_context('paper')
g = sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df)
g.fig.suptitle('Box plot', y = 1.07) # y is for adjusting the height of the title( as by deafult it is quite low(i.e. 1))
```

### 2. assigning a title to AxesSubplot object(scatterplot, countplot, boxplot etc. ) - use g.set_title

```python
g = sns.boxplot(x = 'island', y = 'body_mass_g', data = df)
g.set_title('Box plot', y =1.07)
```

### 3. title for subplots

```python
g = sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter', col = 'species')
g.fig.suptitle('Box plots', y = 1.08) # for naming the entire title (like as on facetgrid)
g.set_titles('This is {col_name}')#naming subplot individually (use set_titles and not set_title which was for axessubplot)
# similar steps could be performed by catplot() also
```

### 4. adding axis label - below code works for both facetgrid and axesubplot objects

```python
g = sns.relplot(x ='bill_length_mm', y = 'bill_depth_mm', data = df, kind = 'scatter')
g.set(xlabel = 'new X label', ylabel = 'new y label') # it will change the original x n y labels set by seaborn
```

### 5. Rotating x-axis tick labels - below code works for both facetgrid and axesubplot objects

```python
g = sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df)
plt.xticks(rotation = 90) #check the xlabel
plt.show()
```

### 6. Setting x and y axis limits

```python
g = sns.catplot(x = 'island', y = 'body_mass_g', kind = 'box', data = df)
# g.set(xlim = (0,10)) # it also could be changed when both the axes have quantitative variables
g.set(ylim = (0,8000))
plt.show()
```
